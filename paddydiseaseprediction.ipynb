```python
# ==========================================
# COMPLETE PADDY DISEASE CLASSIFICATION WORKFLOW
# ==========================================

import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.utils.class_weight import compute_class_weight
import tensorflow as tf
from tensorflow.keras import layers, models, applications, optimizers, losses, regularizers
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint, LearningRateScheduler

# Enable mixed precision training
policy = tf.keras.mixed_precision.Policy('mixed_float16')
tf.keras.mixed_precision.set_global_policy(policy)

class PaddyDiseaseClassifier:
    def __init__(self, img_size=(299, 299), batch_size=32):
        self.img_size = img_size
        self.batch_size = batch_size
        self.class_names = []
        self.model = None
    
    def prepare_data(self, data_dir):
        """Load and prepare training/validation datasets"""
        train_ds = tf.keras.utils.image_dataset_from_directory(
            data_dir,
            validation_split=0.2,
            subset="training",
            seed=123,
            image_size=self.img_size,
            batch_size=self.batch_size,
            label_mode='categorical'
        )
        
        self.class_names = train_ds.class_names
        
        val_ds = tf.keras.utils.image_dataset_from_directory(
            data_dir,
            validation_split=0.2,
            subset="validation",
            seed=123,
            image_size=self.img_size,
            batch_size=self.batch_size,
            label_mode='categorical'
        )
        
        # Optimize datasets
        train_ds = train_ds.cache().prefetch(tf.data.AUTOTUNE)
        val_ds = val_ds.cache().prefetch(tf.data.AUTOTUNE)
        
        return train_ds, val_ds
    
    def create_model(self):
        """Create enhanced InceptionV3-based model"""
        base_model = applications.InceptionV3(
            include_top=False,
            input_shape=(*self.img_size, 3),
            weights="imagenet"
        )
        
        # Freeze base model layers initially
        for layer in base_model.layers:
            layer.trainable = False
        
        # Unfreeze last 50 layers (excluding BatchNorm)
        for layer in base_model.layers[-50:]:
            if not isinstance(layer, layers.BatchNormalization):
                layer.trainable = True

        inputs = layers.Input(shape=(*self.img_size, 3))
        x = applications.inception_v3.preprocess_input(inputs)
        x = base_model(x)
        
        # Enhanced feature extraction
        x1 = layers.GlobalAveragePooling2D()(x)
        x2 = layers.GlobalMaxPooling2D()(x)
        x = layers.concatenate([x1, x2])
        
        # Regularized classifier
        x = layers.Dense(512, activation='swish',
                        kernel_regularizer=regularizers.l1_l2(1e-5, 1e-4))(x)
        x = layers.BatchNormalization()(x)
        x = layers.Dropout(0.5)(x)
        outputs = layers.Dense(len(self.class_names), activation='softmax', dtype='float32')(x)
        
        model = models.Model(inputs, outputs)
        
        optimizer = optimizers.AdamW(learning_rate=1e-4, weight_decay=1e-4)
        model.compile(
            optimizer=optimizer,
            loss=losses.CategoricalCrossentropy(label_smoothing=0.1),
            metrics=['accuracy']
        )
        
        return model
    
    def train(self, train_ds, val_ds, epochs=40):
        """Train the model with callbacks and class weights"""
        # Calculate class weights
        y_train = np.concatenate([y.numpy() for x, y in train_ds], axis=0)
        class_weights = compute_class_weight('balanced', 
                                          classes=np.arange(len(self.class_names)), 
                                          y=np.argmax(y_train, axis=1))
        class_weights = {i: weight for i, weight in enumerate(class_weights)}
        
        # Callbacks
        callbacks = [
            EarlyStopping(patience=10, restore_best_weights=True, monitor='val_accuracy'),
            ReduceLROnPlateau(monitor='val_accuracy', factor=0.5, patience=3, min_lr=1e-6),
            ModelCheckpoint('best_model.keras', monitor='val_accuracy', save_best_only=True)
        ]
        
        history = self.model.fit(
            train_ds,
            validation_data=val_ds,
            epochs=epochs,
            class_weight=class_weights,
            callbacks=callbacks
        )
        
        return history
    
    def evaluate(self, dataset):
        """Evaluate model performance"""
        y_true = []
        y_pred = []
        
        for images, labels in dataset:
            y_true.extend(labels.numpy())
            preds = self.model.predict(images, verbose=0)
            y_pred.extend(preds)
        
        y_true = np.array(y_true)
        y_pred = np.array(y_pred)
        
        # Confusion matrix
        cm = confusion_matrix(np.argmax(y_true, axis=1), np.argmax(y_pred, axis=1))
        plt.figure(figsize=(12, 10))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                    xticklabels=self.class_names,
                    yticklabels=self.class_names)
        plt.title('Confusion Matrix')
        plt.show()
        
        # Classification report
        print(classification_report(
            np.argmax(y_true, axis=1),
            np.argmax(y_pred, axis=1),
            target_names=self.class_names
        ))
    
    def predict(self, input_path):
        """Make predictions on new images"""
        if os.path.isfile(input_path):
            image_paths = [input_path]
        elif os.path.isdir(input_path):
            image_paths = [os.path.join(input_path, f) 
                          for f in os.listdir(input_path) 
                          if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
        else:
            raise ValueError("Input path must be valid file or directory")
        
        results = []
        for img_path in image_paths:
            try:
                img = tf.io.read_file(img_path)
                img = tf.image.decode_jpeg(img, channels=3)
                img = tf.image.resize(img, self.img_size)
                img = applications.inception_v3.preprocess_input(img)
                img = tf.expand_dims(img, axis=0)
                
                pred = self.model.predict(img, verbose=0)
                class_idx = np.argmax(pred)
                confidence = np.max(pred) * 100
                
                results.append({
                    'image_path': img_path,
                    'predicted_class': self.class_names[class_idx],
                    'confidence': confidence
                })
                
            except Exception as e:
                print(f"Error processing {img_path}: {str(e)}")
        
        return pd.DataFrame(results)
    
    def save_model(self, path):
        """Save the trained model"""
        self.model.save(path)
        print(f"Model saved to {path}")
    
    def load_model(self, path):
        """Load a saved model"""
        self.model = tf.keras.models.load_model(path)
        print(f"Model loaded from {path}")

# Main workflow
if __name__ == "__main__":
    # Initialize classifier
    classifier = PaddyDiseaseClassifier()
    
    # Training mode
    train_mode = input("Run in training mode? (y/n): ").lower() == 'y'
    
    if train_mode:
        # 1. Prepare data
        data_dir = "/kaggle/input/paddy-doctor/paddy-disease-classification/train_images"
        train_ds, val_ds = classifier.prepare_data(data_dir)
        print(f"Class names: {classifier.class_names}")
        
        # 2. Create and train model
        classifier.model = classifier.create_model()
        classifier.model.summary()
        print("\nTraining model...")
        history = classifier.train(train_ds, val_ds)
        
        # 3. Evaluate
        print("\nEvaluating model...")
        classifier.evaluate(val_ds)
        
        # 4. Save model
        classifier.save_model('paddy_disease_model.keras')
    else:
        # Prediction mode
        model_path = input("Enter path to trained model: ")
        classifier.load_model(model_path)
        
        input_path = input("Enter path to image or directory: ")
        results = classifier.predict(input_path)
        
        print("\nPrediction Results:")
        print(results)
        
        if len(results) > 1:
            results.to_csv('predictions.csv', index=False)
            print("Predictions saved to predictions.csv")
```

```python
# ==========================================
# PADDY DISEASE CLASSIFICATION - PREDICTION PIPELINE
# ==========================================

import os
import pandas as pd
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras import layers, models, applications

class PaddyDiseasePredictor:
    def __init__(self):
        # Initialize with correct image size (100, 100) based on the error
        self.img_size = (100, 100)  # Adjusted to match model's expected input
        self.batch_size = 32
        self.class_names = [
            'bacterial_leaf_blight', 
            'bacterial_leaf_streak', 
            'bacterial_panicle_blight', 
            'blast', 
            'brown_spot', 
            'dead_heart', 
            'downy_mildew', 
            'hispa', 
            'normal', 
            'tungro'
        ]
        self.model = None
    
    def load_model(self, model_path):
        """Load the trained model with proper error handling"""
        if not os.path.exists(model_path):
            raise FileNotFoundError(f"Model file not found: {model_path}")
        
        # Only accept valid model file extensions
        if not model_path.lower().endswith(('.h5', '.keras')):
            raise ValueError("Model file must be .h5 or .keras format")
        
        print(f"\nLoading model from {model_path}")
        try:
            self.model = tf.keras.models.load_model(model_path)
            print("Model loaded successfully")
            # Verify input shape
            expected_shape = self.model.input_shape[1:3]
            if expected_shape != self.img_size:
                print(f"Note: Model expects input size {expected_shape}, adjusting our processing...")
                self.img_size = expected_shape
        except Exception as e:
            raise ValueError(f"Error loading model: {str(e)}")
    
    def process_input(self, input_path):
        """Handle both directory and single image input"""
        if not os.path.exists(input_path):
            raise FileNotFoundError(f"Path does not exist: {input_path}")
        
        if os.path.isdir(input_path):
            # Directory case
            test_images = [f for f in os.listdir(input_path) 
                         if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
            if not test_images:
                raise ValueError(f"No images found in directory: {input_path}")
            print(f"Found {len(test_images)} images in directory")
            return [os.path.join(input_path, f) for f in test_images]
        
        elif os.path.isfile(input_path):
            # Single image case
            if not input_path.lower().endswith(('.jpg', '.jpeg', '.png')):
                raise ValueError("File must be an image (jpg, jpeg, png)")
            print("Processing single image")
            return [input_path]
        
        raise ValueError(f"Invalid input: {input_path}. Must be directory or image file")
    
    def preprocess_image(self, img_path):
        """Preprocess a single image for model prediction"""
        img = tf.io.read_file(img_path)
        img = tf.image.decode_jpeg(img, channels=3)
        img = tf.image.resize(img, self.img_size)
        img = img / 255.0  # Simple normalization (adjust if your model uses different preprocessing)
        return tf.expand_dims(img, axis=0)  # Add batch dimension
    
    def create_dataset(self, file_paths):
        """Create TensorFlow dataset from file paths"""
        def load_and_preprocess(path):
            img = tf.io.read_file(path)
            img = tf.image.decode_jpeg(img, channels=3)
            img = tf.image.resize(img, self.img_size)
            img = img / 255.0  # Simple normalization
            return img
        
        dataset = tf.data.Dataset.from_tensor_slices(file_paths)
        dataset = dataset.map(load_and_preprocess, num_parallel_calls=tf.data.AUTOTUNE)
        return dataset.batch(self.batch_size).prefetch(tf.data.AUTOTUNE)
    
    def predict(self, input_path):
        """Make predictions on new images"""
        file_paths = self.process_input(input_path)
        
        if len(file_paths) == 1:
            # Single image processing
            processed_img = self.preprocess_image(file_paths[0])
            pred = self.model.predict(processed_img, verbose=0)
        else:
            # Batch processing for multiple images
            dataset = self.create_dataset(file_paths)
            pred = self.model.predict(dataset, verbose=1)
        
        # Process predictions
        predicted_classes = [self.class_names[np.argmax(p)] for p in pred]
        confidences = [np.max(p) * 100 for p in pred]
        
        # Create results DataFrame
        results = pd.DataFrame({
            'image_path': file_paths,
            'predicted_class': predicted_classes,
            'confidence': confidences
        })
        
        return results
    
    def display_results(self, results):
        """Display or save results based on number of images"""
        os.makedirs('output', exist_ok=True)
        
        if len(results) == 1:
            # Single image - display
            row = results.iloc[0]
            print(f"\nRESULT FOR {row['image_path']}:")
            print(f"Predicted class: {row['predicted_class']}")
            print(f"Confidence: {row['confidence']:.2f}%")
            
            # Display the image
            img = plt.imread(row['image_path'])
            plt.figure(figsize=(4, 4))
            plt.imshow(img)
            plt.title(f"Predicted: {row['predicted_class']}\nConfidence: {row['confidence']:.2f}%")
            plt.axis('off')
            plt.show()
        else:
            # Multiple images - save to CSV
            output_path = os.path.join('output', 'predictions.csv')
            results.to_csv(output_path, index=False)
            print(f"\nSaved predictions for {len(results)} images to {output_path}")
            print("Sample predictions:")
            print(results.head())

def main():
    print("=== Paddy Disease Classification Predictor ===")
    predictor = PaddyDiseasePredictor()
    
    try:
        # Get model path from user
        model_path = input("\nEnter path to trained model (.h5 or .keras file): ").strip()
        predictor.load_model(model_path)
        
        # Get input for prediction
        input_path = input("\nEnter path to image or directory: ").strip()
        
        # Make predictions
        results = predictor.predict(input_path)
        
        # Display/save results
        predictor.display_results(results)
        
    except Exception as e:
        print(f"\nERROR: {str(e)}")
        print("Please check your paths and try again.")

if __name__ == "__main__":
    main()
    print("\nPrediction process completed!")
```
